```python
import csv
import pandas as pd
from arango import ArangoClient
```

```python
raw_df = pd.read_csv('../data/raw.csv')
```

```python
sender_df = raw_df[['sender_id', 'sender_name']].copy()
sender_df.rename(columns={'sender_id':'_key', 'sender_name':'name'}, inplace=True)
```

```python
receiver_df = raw_df[['receiver_id', 'receiver_name']].copy()
receiver_df.rename(columns={'receiver_id':'_key', 'receiver_name':'name'}, inplace=True)
```

```python
organization_df = pd.concat([sender_df, receiver_df], ignore_index=True).drop_duplicates(ignore_index=True)
```
